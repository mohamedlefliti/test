```python
# ===================================================
# LCAA Fuel Combustion Simulation
# Low-Carbon Alcohol-Ammonia Fuel with Nanocatalysts
# Calculates: LHV, CO2 emissions, efficiency gain
# Author: Lefliti Mohamed, MSc Computer Science, NUIST
# ===================================================

import numpy as np

# ---------------------------------------------------
# 1. Fuel Component Properties (LHV and Carbon Content)
# Source: NIST Chemistry WebBook, SAE Fuel Database
# ---------------------------------------------------
components = {
    'Ethanol': {
        'LHV_MJ_kg': 26.8,
        'carbon_per_kg_fuel': 0.521,  # kg C per kg pure component
        'mass_fraction': 0.48
    },
    'Methanol': {
        'LHV_MJ_kg': 19.9,
        'carbon_per_kg_fuel': 0.375,
        'mass_fraction': 0.16
    },
    'Isopropanol': {
        'LHV_MJ_kg': 30.1,
        'carbon_per_kg_fuel': 0.600,
        'mass_fraction': 0.08
    },
    'Butanol': {
        'LHV_MJ_kg': 33.0,
        'carbon_per_kg_fuel': 0.648,
        'mass_fraction': 0.08
    },
    'Ammonia': {
        'LHV_MJ_kg': 18.6,
        'carbon_per_kg_fuel': 0.000,  # Carbon-free
        'mass_fraction': 0.20
    }
}

# ---------------------------------------------------
# 2. Calculate Total LHV and Total Carbon Content
# ---------------------------------------------------
total_LHV = 0.0
total_carbon_per_kg_fuel = 0.0

print("Fuel Composition Analysis by Mass:")
print("-" * 70)
print(f"{'Component':<15} {'Mass Fraction':<14} {'LHV Contribution':<18} {'Carbon (kg/kg)':<16}")
print("-" * 70)

for name, data in components.items():
    lhv_contribution = data['LHV_MJ_kg'] * data['mass_fraction']
    carbon_kg = data['carbon_per_kg_fuel'] * data['mass_fraction']
    total_LHV += lhv_contribution
    total_carbon_per_kg_fuel += carbon_kg
    print(f"{name:<15} {data['mass_fraction']:<14.2f} "
          f"{lhv_contribution:<18.2f} {carbon_kg:<16.3f}")

print("-" * 70)
print(f"{'Total':<15} "
      f"{'—':<14} "
      f"LHV = {total_LHV:.2f} MJ/kg "
      f" | Carbon = {total_carbon_per_kg_fuel:.3f} kg/kg fuel")
print()

# ---------------------------------------------------
# 3. CO2 Emissions Calculation
# Reaction: C + O2 → CO2 → 44g CO2 / 12g C = 3.67 kg CO2 per kg C
# ---------------------------------------------------
CO2_PER_KG_CARBON = 44 / 12  # ≈ 3.67
CO2_kg_per_kg_fuel = total_carbon_per_kg_fuel * CO2_PER_KG_CARBON

# Fuel density (from your specification)
density_g_per_cm3 = 0.778  # g/cm³
density_kg_per_L = density_g_per_cm3  # since 1 cm³ = 1 mL

CO2_kg_per_L = CO2_kg_per_kg_fuel * density_kg_per_L

print("CO2 Emissions:")
print(f"  - {CO2_kg_per_kg_fuel:.3f} kg CO2 / kg of fuel")
print(f"  - {CO2_kg_per_L:.3f} kg CO2 / liter of fuel")
print()

# ---------------------------------------------------
# 4. Comparison with Gasoline
# ---------------------------------------------------
gasoline_LHV_MJ_kg = 42.5
gasoline_density_kg_per_L = 0.74
gasoline_carbon_per_kg = 0.87  # kg C per kg gasoline
gasoline_CO2_kg_per_kg = gasoline_carbon_per_kg * CO2_PER_KG_CARBON
gasoline_CO2_kg_per_L = gasoline_CO2_kg_per_kg * gasoline_density_kg_per_L

reduction_vs_gasoline = ((gasoline_CO2_kg_per_L - CO2_kg_per_L) / gasoline_CO2_kg_per_L) * 100

print("Comparison with Gasoline:")
print(f"  - Gasoline: {gasoline_CO2_kg_per_L:.3f} kg CO2/L")
print(f"  - LCAA Fuel: {CO2_kg_per_L:.3f} kg CO2/L")
print(f"  - CO2 Reduction: {reduction_vs_gasoline:.1f}%")
print()

# ---------------------------------------------------
# 5. Effect of Nanocatalysts on Combustion Efficiency
# ---------------------------------------------------
base_efficiency = 0.95        # 95% without catalysts
enhanced_efficiency = 0.98    # 98% with TiO2/ZnO nanoparticles

effective_LHV_base = total_LHV * base_efficiency
effective_LHV_enhanced = total_LHV * enhanced_efficiency
efficiency_gain_percent = ((effective_LHV_enhanced - effective_LHV_base) / effective_LHV_base) * 100

print("Nanocatalyst Impact on Combustion Efficiency:")
print(f"  - Effective LHV (without nanoparticles): {effective_LHV_base:.2f} MJ/kg")
print(f"  - Effective LHV (with nanoparticles):    {effective_LHV_enhanced:.2f} MJ/kg")
print(f"  - Improvement in usable energy: +{efficiency_gain_percent:.1f}%")
print()

# ---------------------------------------------------
# 6. Final Summary
# ---------------------------------------------------
print("=" * 70)
print("                    LCAA FUEL SIMULATION SUMMARY")
print("=" * 70)
print(f"• Lower Heating Value (LHV):         {total_LHV:.2f} MJ/kg")
print(f"• Density:                           {density_kg_per_L:.3f} kg/L")
print(f"• CO2 Emissions:                     {CO2_kg_per_L:.3f} kg/L")
print(f"• CO2 Reduction vs Gasoline:         {reduction_vs_gasoline:.1f}%")
print(f"• Combustion Efficiency (with nano): {enhanced_efficiency:.0%}")
print(f"• Effective Energy Output:           {effective_LHV_enhanced:.2f} MJ/kg")
print(f"• Engine Compatibility:              Yes (drop-in ready)")
print(f"• Physical State:                    Liquid at ambient conditions")
print("=" * 70)
```
